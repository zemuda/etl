```python
changes_made = False
# iterate through each column to check for object types that can be converted to datetime
for col in result_summarized_table.columns:
    if result_summarized_table[col].dtype == "object":
        try:
            # Check if column contains date-like strings
            sample = result_summarized_table[col].dropna().head(10)
            if not sample.empty:
                # Try to convert to datetime
                pd.to_datetime(sample, errors="raised")
                result_summarized_table[col] = pd.to_datetime(result_summarized_table[col], errors="coerce")
                changes_made = True           
        except:
            pass
```

```python
# Identify all numeric columns
object_cols = result_detailed_table.select_dtypes(include=["object"]).columns

# Create a boolean mask to find rows with missing values in numeric columns
# The `any(axis=1)` checks if ANY column in that row has a missing value
missing_numeric_rows_mask = result_detailed_table[object_cols].isnull().any(axis=1)

# Use the mask to select and display the entire rows
rows_with_missing_numeric = result_detailed_table[missing_numeric_rows_mask]

print("\nRows with missing values in numeric columns: " )
print(missing_numeric_rows_mask.column)
```

```python
# Select only the numeric columns
numeric_cols = result_detailed_table.select_dtypes(include=np.number)

# Create a boolean mask where a row has at least one negative value
rows_with_any_negative = (numeric_cols < 0).any(axis=1)

# Filter the original DataFrame using the boolean mask
result_df = result_detailed_table[rows_with_any_negative]

print("\nRows with any negative values in numeric columns:")
result_df.head()
```

```python
# result_detailed_table drop columns if has a lot of null values
# Calculate the percentage of missing values for each column
null_percentages = result_detailed_table.isnull().mean() * 100

# Set a threshold (e.g., drop columns with more than 50% missing values)
threshold = 1

# Get the list of columns to drop
columns_to_drop = null_percentages[null_percentages > threshold].index.tolist()

# Drop the columns from the DataFrame
result_detailed_table.drop(columns=columns_to_drop, inplace=True)

print("Columns dropped:", columns_to_drop)
print("\nDataFrame after dropping columns:")
result_detailed_table.isnull().sum()
```

```python
# result_summarized_table drop columns if has a lot of null values
# Calculate the percentage of missing values for each column
null_percentages = result_summarized_table.isnull().mean() * 100

# Set a threshold (e.g., drop columns with more than 50% missing values)
threshold = 1

# Get the list of columns to drop
columns_to_drop = null_percentages[null_percentages > threshold].index.tolist()

# Drop the columns from the DataFrame
result_summarized_table.drop(columns=columns_to_drop, inplace=True)

print("Columns dropped:", columns_to_drop)
print("\nDataFrame after dropping columns:")
result_summarized_table.isnull().sum()
```

```python
# if result_detailed_table['VAT'] == 0.0 and result_detailed_table['DISC'] == 0.0 then drop the columns
# Check if ALL values in 'VAT' and 'DISC' columns are 0.0
vat_is_zero = (result_detailed_table['VAT'] == 0.0).all()
disc_is_zero = (result_detailed_table['DISC'] == 0.0).all()

# Check the combined condition
if vat_is_zero and disc_is_zero:
    # Drop the columns if the condition is met
    result_detailed_table = result_detailed_table.drop(columns=['VAT', 'DISC'])
    print("\n'VAT' and 'DISC' columns dropped because all their values were 0.0.")
else:
    print("\nCondition not met. Columns 'VAT' and 'DISC' were not dropped.")

print("\nResulting DataFrame:")
result_detailed_table.dtypes
```
